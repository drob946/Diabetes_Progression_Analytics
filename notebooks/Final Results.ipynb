### Final Results
- This notebook consolidates the key findings from the analysis, comparing models and evaluating their performance.



```python
import pandas as pd
import matplotlib.pyplot as plt
model_metrics = pd.DataFrame({
    "Model": ["Linear Regression", "Random Forest", "Gradient Boosting"],
    "MAE": [42.79, 43.78, 42.84],
    "MSE": [2900.19, 2975.17, 2886.36],
    "R²": [0.45, 0.44, 0.46]
})

model_metrics
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=False)


model_metrics.plot(x="Model", y="MAE", kind="bar", ax=axes[0], color="gray", edgecolor="black", legend=False)
axes[0].set_title("Mean Absolute Error (MAE)", fontsize=14)
axes[0].set_ylabel("MAE", fontsize=12)
axes[0].set_xlabel("Model", fontsize=12)
axes[0].tick_params(axis="x", labelsize=10)
axes[0].grid(axis="y", linestyle="--", linewidth=0.5, alpha=0.7)


model_metrics.plot(x="Model", y="MSE", kind="bar", ax=axes[1], color="steelblue", edgecolor="black", legend=False)
axes[1].set_title("Mean Squared Error (MSE)", fontsize=14)
axes[1].set_ylabel("MSE", fontsize=12)
axes[1].set_xlabel("Model", fontsize=12)
axes[1].tick_params(axis="x", labelsize=10)
axes[1].grid(axis="y", linestyle="--", linewidth=0.5, alpha=0.7)


model_metrics.plot(x="Model", y="R²", kind="bar", ax=axes[2], color="indianred", edgecolor="black", legend=False)
axes[2].set_title("R² Score", fontsize=14)
axes[2].set_ylabel("R²", fontsize=12)
axes[2].set_xlabel("Model", fontsize=12)
axes[2].tick_params(axis="x", labelsize=10)
axes[2].grid(axis="y", linestyle="--", linewidth=0.5, alpha=0.7)


plt.tight_layout()
plt.show()
```

### Key Findings
#### Mean Absolute Error (MAE)
- MAE measures the average magnitude of the errors between the predicted and actual values, without considering their direction. It's in the same units as the target variable.

##### Observation
- All three models (Linear Regression, Random Forest, and Gradient Boosting) have similar MAE values, indicating they all perform similarly in terms of the average prediction error.
- Gradient Boosting has a slightly lower MAE, which suggests it makes smaller errors on average compared to the other models.

#### Mean Squared Error (MSE)
- MSE measures the average squared difference between predicted and actual values, penalizing larger errors more than smaller ones.
##### Observation
- Gradient Boosting achieves the lowest MSE, reflecting its ability to minimize large prediction errors.
- Linear Regression and Random Forest have slightly higher MSE, indicating they struggle a bit more with outliers or larger deviations.


#### R² Score
- The R² score represents the proportion of variance in the target variable explained by the model. It ranges from 0 to 1, with 1 being a perfect fit.
##### Observation
- Gradient Boosting has the highest R² score, meaning it explains the most variance in the target variable.
- All three models have similar R² scores (~0.45), reinforcing the idea that the dataset limits the models' potential performance.

### Best Model
- **Gradient Boosting** achieved the best performance with the **highest R² (0.46)** and **lowest MSE (2886.36)**, indicating it captured more of the variance in the target variable. However, **all models performed similarly due to the dataset's limitations**, such as sample size and feature set.
