```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Load the processed dataset
processed_path = "../data/processed/diabetes_scaled.csv"
scaled_data = pd.read_csv(processed_path)

# Check the first few rows
print(scaled_data.head())
```

```python
# Summarize the data with descriptive statistics
print("Descriptive Statistics:")
print(scaled_data.describe())

# Check for missing values
print("\nMissing Values:")
print(scaled_data.isnull().sum())

# Count unique values for each feature
print("\nUnique Values in Each Feature:")
for col in scaled_data.columns:
    print(f"{col}: {scaled_data[col].nunique()} unique values")
```

```python
# Plot histograms for numerical features
scaled_data.hist(figsize=(12, 10), bins=15, edgecolor='black')
plt.suptitle("Feature Distributions", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Heatmap of correlations
plt.figure(figsize=(10, 8))
sns.heatmap(scaled_data.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Pairplot for selected features
selected_features = ['bmi', 's1', 's2', 's3', 'target']
sns.pairplot(scaled_data[selected_features])
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.show()
```

```python
# Boxplot for all numerical features
plt.figure(figsize=(15, 10))
sns.boxplot(data=scaled_data.iloc[:, :-1])  # Exclude the target column
plt.xticks(rotation=45)
plt.title("Boxplots of Features")
plt.show()
```
